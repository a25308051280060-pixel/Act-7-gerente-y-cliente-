```python
# --- SIMULADOR DE TIENDA INTELIGENTE v1.0 ---

# Inicialización de variables base (Datos del sistema)
nombre_tienda = "Prepa-Store Inteligente"
precio_unitario = 150.0  # Precio base de un artículo
descuento_gerente = 0.0  # Configurado manualmente por el gerente
ejecutar_sistema = "si"  # Variable de control para el bucle principal [4]

print("--- BIENVENIDO AL SISTEMA DE GESTIÓN IA: ", nombre_tienda, " ---")

# Bucle principal interactivo utilizando 'while' [4, 5]
while ejecutar_sistema == "si":
    print("\n******************************************")
    print("MENÚ DE INTERACCIÓN:")
    print("1. Modo Gerente (Configurar Reglas)")
    print("2. Modo Cliente (Simular Compra con IA)")
    print("3. Salir del Sistema")
    print("******************************************")

    opcion = input("Seleccione una opción (1, 2 o 3): ")

    # ESTRUCTURA DE CONTROL SELECTIVA MÚLTIPLE (if-elif-else) [6]
    if opcion == "1":
        # --- LÓGICA DEL GERENTE ---
        print("\n[MODO GERENTE ACTIVADO]")
        print("Actualmente el descuento base es del:", descuento_gerente * 100, "%")

        # Modified prompt for clarity
        nuevo_desc = float(input("Ingrese el nuevo descuento base como un valor decimal entre 0 y 1 (ejemplo: 0.1 para 10%): "))

        # Uso de operadores relacionales para validar la entrada [7]
        if nuevo_desc > 0.5:
            print("IA Alerta: ¡Atención! Un descuento mayor al 50% podría generar pérdidas.")
        # Add a check to ensure discount is not above 1
        if nuevo_desc > 1.0:
            print("IA Alerta: El descuento no puede ser mayor a 1 (100%). Se establecerá en 1 (100%).")
            descuento_gerente = 1.0
        elif nuevo_desc < 0:
            print("IA Alerta: El descuento no puede ser negativo. Se establecerá en 0.")
            descuento_gerente = 0.0
        else:
            descuento_gerente = nuevo_desc
        print("Configuración guardada exitosamente.")

    elif opcion == "2":
        # --- LÓGICA DEL CLIENTE (SIMULACIÓN IA) ---
        print("\n[MODO CLIENTE ACTIVADO]")
        print("El producto estrella cuesta: $", precio_unitario)

        cantidad = int(input("¿Cuántas unidades desea comprar? "))
        presupuesto = float(input("¿De cuánto dinero dispone para esta compra? ($): "))
        es_alumno = input("¿Es alumno de esta preparatoria? (si/no): ")

        # Cálculo aritmético básico [8]
        subtotal = precio_unitario * cantidad

        # PROCESO DE DECISIÓN IA (Uso de operadores lógicos y relacionales) [7, 9]
        # La IA decide un "Bono de Lealtad" extra
        if cantidad >= 10 and es_alumno == "si":
            bono_ia = 0.10  # 10% extra por gran volumen y ser alumno
            print("IA Decision: Detectado cliente VIP escolar. Aplicando bono de lealtad.")
        elif es_alumno == "si" or cantidad >= 5:
            bono_ia = 0.05  # 5% extra si cumple una de las dos condiciones [10]
        else:
            bono_ia = 0.0

        # Cálculo del total final con todos los descuentos acumulados
        descuento_total = descuento_gerente + bono_ia

        # Ensure total discount does not exceed 1 (100%)
        if descuento_total > 1.0:
            descuento_total = 1.0
            print("IA Alerta: El descuento total acumulado excede 100%. Se ha limitado a 100%.")

        total_pagar = subtotal * (1 - descuento_total)

        print("\n--- ANÁLISIS DE LA IA ---")
        print("Subtotal sin ofertas: $", subtotal)
        print("Descuento total aplicado (IA + Gerente): ", descuento_total * 100, "%")
        print("Total final calculado: $", total_pagar)

        # La IA determina si la transacción es posible [3, 7]
        if presupuesto >= total_pagar:
            cambio = presupuesto - total_pagar # Operador de resta [8]
            print("IA Resultado: Transacción ACEPTADA. Su cambio es: $", cambio)
        else:
            faltante = total_pagar - presupuesto
            print("IA Resultado: Transacción RECHAZADA. Le faltan $", faltante)

    elif opcion == "3":
        # Opción para romper el bucle y salir [5, 11]
        ejecutar_sistema = "no"
        print("IA: Desconectando sistema de gestión. Tenga un excelente día.")

    else:
        # Manejo de opciones no válidas [6]
        print("Error: La opción seleccionada no existe en el menú.")

print("\n--- PROGRAMA FINALIZADO ---")
print("Programa terminado por Ambar Fierro")
```
